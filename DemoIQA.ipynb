```python
import cv2
import numpy as np

def check_image_quality(image_path):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # 1. Kiểm tra độ mờ (Blurriness) bằng biến thể Laplacian
    blur_score = cv2.Laplacian(img, cv2.CV_64F).var()

    # 2. Kiểm tra độ nhiễu (Ước tính đơn giản qua độ lệch chuẩn vùng nền)
    noise_score = np.std(img)

    # 3. Thiết lập ngưỡng (Thresholds) - Các số này bạn cần tinh chỉnh sau khi test dữ liệu thật
    BLUR_THRESHOLD = 100.0  # Dưới mức này là quá mờ
    NOISE_UPPER_LIMIT = 80.0 # Trên mức này là quá nhiễu hạt

    is_qualified = True
    reason = "Đạt tiêu chuẩn"

    if blur_score < BLUR_THRESHOLD:
        is_qualified = False
        reason = "Ảnh quá mờ"
    elif noise_score > NOISE_UPPER_LIMIT:
        is_qualified = False
        reason = "Ảnh quá nhiễu"

    return is_qualified, blur_score, reason

# Thử nghiệm
qualified, score, msg = check_image_quality('/content/Thiết kế chưa có tên (1).png')
if qualified:
    print(f"Tiến hành xử lý SwinIR... (Score: {score})")
else:
    print(f"Từ chối ảnh: {msg}")
```
